```python
#Below is the sample sql file
from sqlite3 import connect
conn = connect(':memory:')
df3 = pd.DataFrame(data=[[0, '10/11/12'], [1, '12/11/10']],
                  columns=['int_column', 'date_column'])
df3.to_sql('test_data', conn)


#df4.head()
```
